## Measles Cases by WHO Region for 2018 and 2019

```python
measles_3 = pd.read_csv('big_measlesregionalsummary.csv', header=0)
```

```python
measles_3
```

```python
list_2018 = list(measles_3['Total measles (2018)'])
```

```python
list_2019 = list(measles_3['Total measles (2019)'])
```

```python
percent_list = []
```

```python
for x in range(len(list_2018)):
    try:
        percent_list.append((int(list_2019[x])/int(list_2018[x]))*100)
        
    except: ValueError
```

```python
percent_list
```

```python
measles_3['Percent Increase 2018-2019(Total Measles)'] = percent_list
```

```python
WHO_MeaslesIncrease = measles_3
WHO_MeaslesIncrease
```

```python
WHO_MeaslesPercent = WHO_MeaslesIncrease[['WHO Region', 'Percent Increase 2018-2019(Total Measles)']]
```

```python
WHO_MeaslesPercent
```

```python
WHO_Region_MCV1_MCV2_2018_Cleaned1 = WHO_Region_MCV1_MCV2_2018.drop(0)
WHO_Region_MCV1_MCV2_2018_Cleaned2 = WHO_Region_MCV1_MCV2_2018_Cleaned1.reindex([2,3,6,5,4,7,1])
WHO_Region_MCV1_MCV2_2018_Cleaned2.reset_index()
```

```python
WHO_Region_MCV1_MCV2_2018_Cleaned2
```

## Merge these tables to evaluate importance of vaccination correlation to measles incidence

```python
m_table1 = pd.merge(wiki_table, WHO_MeaslesPercent, on='index')
m_table1[['WHO-Region', '1980', '1990', '2000', '2005', '2014', 'Percent Increase 2018-2019(Total Measles)']]
```

```python
list1 = list(WHO_Region_MCV1_MCV2_2018_Cleaned2['MCV1 Coverage, 1-Year-Olds (%)'])
list1
```

```python
list2 = list(WHO_Region_MCV1_MCV2_2018_Cleaned2['MCV2 Coverage, 1-Year-Olds (%)'])
list2
```

```python
m_table1['MCV1 Coverage, 1-Year-Olds (%)'] = list1
m_table1['MCV2 Coverage, 1-Year-Olds (%)'] = list2
m_table1
```

```python
Measles_Correlation_Table = m_table1[['index', 'WHO-Region', '1980', '1990', '2000', '2005', '2014', 'Percent Increase 2018-2019(Total Measles)', 'MCV1 Coverage, 1-Year-Olds (%)', 'MCV2 Coverage, 1-Year-Olds (%)' ]]
```

```python
Measles_Correlation_Table
```
